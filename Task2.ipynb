```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r"D:\data\internship\5\Task 2\Sales.xlsx")
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='b')

# 3. Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales (in USD)')
plt.title('Company Sales Trend Over the Year')

# 4. Rotate x-axis labels for readability
plt.xticks(rotation=45)

# 5. Add grid for better visuals
plt.grid(True)

# 6. Show the chart
plt.tight_layout()
plt.show()
```
